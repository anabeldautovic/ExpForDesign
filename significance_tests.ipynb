```python

#Get the results of the paper
paper = pd.read_csv('paper_results.csv')
paper = paper.rename(columns=mapping)
paper = paper.groupby(["Topics", "Dataset"]).sum()
paper
```

```python
#Compute t-statistics
t_test = (means-paper)/(sds/math.sqrt(3))
t_test
```

```python
#Where the t-statistics is greater than 1.96 - the difference is significant?
t_value = 2.92 # df = 2, alpha = 0.05 / look slides about statistics page 21
significant = abs(t_test) > t_value
significant
#only two results is significantly different from the paper results
```
